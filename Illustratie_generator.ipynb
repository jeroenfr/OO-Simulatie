```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
#Vraag 1 (en 2)
slaapduur= np.arange(2,11.5,0.5)
aantalStudenten = np.array([0,0,1,0,0,0,2,4,7,10,16,10,7,5,4,2,1,0,0])
fig, ax = plt.subplots(layout='tight')
ax.bar(x=slaapduur,height=aantalStudenten,width = 0.5,edgecolor='black')
ax.set_xticks(np.arange(2,12))
ax.set_xlabel('Aantal uren slaap')
ax.set_yticks(np.arange(5,25,5))
ax.set_ylabel('Aantal studenten')
fig.savefig('Vraag1.png')
```

```python
#Vraag 2 (A)
results = np.repeat(a=slaapduur,repeats=aantalStudenten)
fig, ax = plt.subplots(figsize=(7,1.5),layout='tight')
ax.boxplot(results,vert=False,widths=0.5,medianprops={"color": "black"},flierprops = {'markerfacecolor':'black'})
ax.set_xticks(np.arange(2,12))
ax.set_yticks([])
fig.savefig('Vraag2A.png',dpi=300, bbox_inches = "tight")
```

```python
def my_formatter(x, pos):
    """Format 1 as 1, 0 as 0, and all values whose absolute values is between
    0 and 1 without the leading "0." (e.g., 0.7 is formatted as .7 and -0.4 is
    formatted as -.4)."""
    val_str = '{:.2f}'.format(x)
    if np.abs(x) > 0 and np.abs(x) < 1:
        return val_str.replace("0", "", 1)
    else:
        return val_str
# Set up the formatter.
major_formatter = FuncFormatter(my_formatter)
```

```python
#Vraag 6
data = {'BH':0.305, 'HA':0.229, 'JS':0.281, 'TC':0.097, 'MM':0.167, 'GV':0.333, 'RC':0.085,
        'SU':0.270, 'DH':0.136, 'TO':0.218, 'RL':0.267, 'JB':0.270, 'WG':0.054, 'MH':0.108,
        'BC':0.301, 'AA':0.143, 'HK':0.341, 'RS':0.261, 'CR':0.115, 'MD':0.125}
data_namen = np.array(list(data.keys()))
data_proporties = np.array(list(data.values()))
fig, ax = plt.subplots(figsize=(7,2),layout='tight')
ax.bar(x=data_namen,height=data_proporties,width = 1,edgecolor='black')
ax.tick_params(axis='x', which='both', bottom=False, top=False);  # Verwijder de x-tick marks
ax.xaxis.set_major_locator(plt.FixedLocator(range(len(data_namen)))); # Gebruik FixedLocator om de x-tick-locaties expliciet op te geven
ax.set_xticklabels(data_namen); # Voeg de x-tick labels toe
ax.yaxis.set_major_formatter(major_formatter)
ax.set_yticks(np.arange(0.05,0.40,0.05));
ax.set_ylabel('A',rotation=0, fontsize=20, labelpad=20);
fig.savefig('Vraag6(A).png',dpi=300, bbox_inches = "tight")
```

```python
data_namen = np.array(['WG','TC','CR','DH','MM','HA','RS','JB','JS','BH','HK','GV','BC','SU','RL','TO','AA','MD','MH','RC'])
gesorteerde_data = {k: data[k] for k in data_namen}
data_namen = np.array(list(gesorteerde_data.keys()))
data_proporties = np.array(list(gesorteerde_data.values()))
fig, ax = plt.subplots(figsize=(7,2),layout='tight')
ax.bar(x=data_namen,height=data_proporties,width = 1,edgecolor='black')
ax.tick_params(axis='x', which='both', bottom=False, top=False);  # Verwijder de x-tick marks
ax.xaxis.set_major_locator(plt.FixedLocator(range(len(data_namen)))); # Gebruik FixedLocator om de x-tick-locaties expliciet op te geven
ax.set_xticklabels(data_namen); # Voeg de x-tick labels toe
ax.yaxis.set_major_formatter(major_formatter)
ax.set_yticks(np.arange(0.05,0.40,0.05));
ax.set_ylabel('B',rotation=0, fontsize=20, labelpad=20);
fig.savefig('Vraag6(B).png',dpi=300, bbox_inches = "tight")
```
